# Word-to-Index Mapping

When we convert documents to vectors, we get a document-term matrix



How do we know which colum corresponds to which word?
It depends: it's up to to us to decide how it is built and how the mapping is made.

## Example:



The vocabulary size $V=5$ (there are 5 unique words), therefore, each word will occupy columns 0 to 4.

## Implementation:
Example of code that could implement a mapping is done in the order new words are discovered:
```
current_index = 0
word2index = {}
for doc in documents:
    tokens = word_tokenize(doc)
    for token in tokens:
        if token not in word2index:
            word2index[token] = current_index
            current_index += 1

```

## Train vs Test

What to do with words in the test set but not the train set ?
* Method 1: ignore them, because the ML model will be trained on the train set and wouldn't know what to do with other words.
* Method 2: create a special index for rare/unknown words. In fact we could assign rare words that fall below a frequency threshold to the unknown index.
